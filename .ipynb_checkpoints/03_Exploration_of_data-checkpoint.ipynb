# Library

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import random
```

# Explore data

```python
images_dir = os.path.expanduser('~/work/Neural_Networks_for_Water_Body_Segmentation/Water_Bodies_Dataset/Images')
masks_dir = os.path.expanduser('~/work/Neural_Networks_for_Water_Body_Segmentation/Water_Bodies_Dataset/Masks')

image_files = sorted([f for f in os.listdir(images_dir) if f.endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff'))])
mask_files = sorted([f for f in os.listdir(masks_dir) if f.endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff'))])

print(f"Number of images: {len(image_files)}")
print(f"Number of masks: {len(mask_files)}")
```

```python

# Function to display an image and its mask
def show_image_and_mask(img_path, mask_path):
    img = Image.open(img_path)
    mask = Image.open(mask_path)
    
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.title("Satellite Image")
    plt.imshow(img)
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.title("Water Mask")
    plt.imshow(mask, cmap='gray')
    plt.axis('off')

    plt.show()

# Display a few examples
for i in range(3):
    img_path = os.path.join(images_dir, image_files[i])
    mask_path = os.path.join(masks_dir, mask_files[i])
    show_image_and_mask(img_path, mask_path)
```

```python
# Simple statistics on image sizes
widths, heights = [], []
for img_file in image_files:
    img = Image.open(os.path.join(images_dir, img_file))
    widths.append(img.width)
    heights.append(img.height)

print(f"Average image size: {int(sum(widths)/len(widths))} x {int(sum(heights)/len(heights))}")
print(f"Minimum size: {min(widths)} x {min(heights)}")
print(f"Maximum size: {max(widths)} x {max(heights)}")
```

```python
# Dataset Balance (water vs non-water pixels) :
# We look at masked images, and see wich pixels are white (representing water), and wich are not.
# We compute the ratio of white pixels => it gives us the part of water in the image

water_pixel_ratios = []
for mask_file in mask_files:
    mask = Image.open(os.path.join(masks_dir, mask_file)).convert('L')
    mask_np = np.array(mask)
    water_pixels = np.sum(mask_np > 128)  # assuming water pixels are white (>128)
    total_pixels = mask_np.size
    water_pixel_ratios.append(water_pixels / total_pixels)

print(f"\nAverage water pixel ratio: {np.mean(water_pixel_ratios):.4f}")
print(f"Min water pixel ratio: {np.min(water_pixel_ratios):.4f}")
print(f"Max water pixel ratio: {np.max(water_pixel_ratios):.4f}")

plt.hist(water_pixel_ratios, bins=20, color='blue', alpha=0.7)
plt.title("Distribution of Water Pixel Ratios in Masks")
plt.xlabel("Water Pixel Ratio")
plt.ylabel("Number of Masks")
plt.show()
```

```python
# Image pixel intensity distribution (brightness)

brightness_means = []
for img_file in image_files:
    img = Image.open(os.path.join(images_dir, img_file)).convert('L')
    brightness_means.append(np.mean(np.array(img)))

plt.hist(brightness_means, bins=30, color='green', alpha=0.7)
plt.title("Distribution of Image Brightness")
plt.xlabel("Mean Brightness")
plt.ylabel("Number of Images")
plt.show()
```

```python
# Preview simple data augmentations

def augment_image_and_mask(img, mask):
    # Flip horizontally with 50% chance
    if random.random() > 0.5:
        img = img.transpose(Image.FLIP_LEFT_RIGHT)
        mask = mask.transpose(Image.FLIP_LEFT_RIGHT)
    # Flip vertically with 50% chance
    if random.random() > 0.5:
        img = img.transpose(Image.FLIP_TOP_BOTTOM)
        mask = mask.transpose(Image.FLIP_TOP_BOTTOM)
    # Rotate 90 degrees randomly
    if random.random() > 0.5:
        img = img.rotate(90, expand=True)
        mask = mask.rotate(90, expand=True)
    return img, mask

print("\nShowing original and augmented image/mask pairs:")
for i in range(2):
    img_path = os.path.join(images_dir, image_files[i])
    mask_path = os.path.join(masks_dir, mask_files[i])
    img = Image.open(img_path)
    mask = Image.open(mask_path)
    img_aug, mask_aug = augment_image_and_mask(img, mask)

    plt.figure(figsize=(12,6))
    plt.subplot(2,2,1)
    plt.title("Original Image")
    plt.imshow(img)
    plt.axis('off')

    plt.subplot(2,2,2)
    plt.title("Original Mask")
    plt.imshow(mask, cmap='gray')
    plt.axis('off')

    plt.subplot(2,2,3)
    plt.title("Augmented Image")
    plt.imshow(img_aug)
    plt.axis('off')

    plt.subplot(2,2,4)
    plt.title("Augmented Mask")
    plt.imshow(mask_aug, cmap='gray')
    plt.axis('off')

    plt.show()
```

```python
empty_masks = []
full_masks = []

for mask_file in mask_files:
    mask = Image.open(os.path.join(masks_dir, mask_file)).convert('L')
    mask_np = np.array(mask)
    
    # Calculate water pixel ratio
    water_ratio = np.sum(mask_np > 128) / mask_np.size
    water_pct = round(water_ratio * 100, 2)
    
    # Track mask types
    if water_ratio == 0:
        empty_masks.append((mask_file, water_pct))
    elif water_ratio == 1:
        full_masks.append((mask_file, water_pct))

# Summary
print(f"\nTotal empty masks (0% water): {len(empty_masks)}")
print(f"Total full masks (100% water): {len(full_masks)}")

print("\n Empty masks:")
for name, pct in empty_masks:
    print(f"  - {name}")

print("\nFull masks:")
for name, pct in full_masks:
    print(f"  - {name}")
```
